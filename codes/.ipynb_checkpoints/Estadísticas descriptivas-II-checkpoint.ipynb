```python
base = open(r'../base_datos/sexo_dni.csv', 'rb').read()
det = chardet.detect(base)
charenc = det['encoding']

sex_dni = pd.read_csv( r'../base_datos/sexo_dni.csv', encoding = charenc)
sex_dni 
```

```python
base_educacion = pd.read_csv( r'../outcomes/base_completa_panel.csv')
base_educacion
```

```python
## Base de sexo y base_educaciÃ³n junta

base_educacion = pd.merge(base_educacion, sex_dni, how="left", on="dni")
base_educacion['sexo'].isnull().sum() # No hay missing values. El gÃ©nero se asignÃ³ a todos los alumnos 
base_educacion
```

```python
# Cantidad de cursos desaprobados y desertados por alumnno en cada periodo

base = base_educacion[~base_educacion['id'].isna()]
base = base.iloc[: :,1:269] # seleccionando columnas 
base['Cursos desaprobados'] = ((base > 0) & (base < 12.5)).sum(axis=1)
base['Desercion de cursos'] = (base.iloc[: :,3:266] == 0).sum(axis=1)
base = base[['dni','period','Cursos desaprobados','Desercion de cursos']]
base
```

```python
# Merge educacion_base + base de nÂ° de cursos desaprobados y deserciÃ³n de cursos

base_educacion = pd.merge(base_educacion,base, how = 'left', on = ['dni','period'])
base_educacion
```

```python
# renombrando las varaibles 

base_educacion = base_educacion.rename({'period':'Periodo', 'section':'Nivel', 'carrera':'Carrera','shift':'Turno'
                                       , 'ave_total':'Nota promedio general', 'ave_gc':'Nota promedio de cursos generales',
                                        'ave_fc':'Nota promedio de cursos de carrera','rate_approv':'ratio de aprobaciÃ³n por alumno'}
                                      , axis=1)
```

```python
# Base de datos estÃ¡ restringida para alumnos los niveles II, IV en el semestre 2019 -2, y III,V en 2020-1

base_educacion = base_educacion[(base_educacion.Nivel > 1) & (base_educacion.Nivel < 6)]
base_educacion = base_educacion.drop(base_educacion[(base_educacion.Periodo == 1) & (base_educacion.Nivel == 5)].index)
```

# Nuevo  - test de diferencia de medias

```python
base = base_educacion[~base_educacion['Nota promedio general'].isna()]
    
print(colored("T-test mean","cyan", attrs=["bold"]),"\n")
t_value, t_test_p = stats.ttest_ind(base["Nota promedio general"][(base.sexo == 1) & (base.Periodo == 2)], 
                                     base["Nota promedio general"][(base.sexo == 0) & (base.Periodo == 2)], equal_var=False)
print('t_value = %.3f, t_test_p = %.3f' % (t_value, t_test_p),"\n")
```

```python
#netflix.groupby( [ 'type' ] )[['num_ctry', 'num_director']].agg( ['mean'] ).reset_index( level = [ 'type'] )

base.groupby(['sexo'])[['Nota promedio general']].agg('mean').reset_index( level = ['sexo'])

# Average mensÂ´ grades: 12.7
# Average women's grades : 13.5 
```

### Se rechaza la hipÃ³tesis nula de la igualdad de la nota promedio entre varones y mujeres al 95% de nivel de significancia

# Average Q. desertion course

```python
base = base_educacion[~base_educacion['Desercion de cursos'].isna()]
base = base[(base['Desercion de cursos']>0) & (base.Periodo == 2)]

print(colored("T-test mean","cyan", attrs=["bold"]),"\n")
t_value, t_test_p = stats.ttest_ind(base["Desercion de cursos"][(base.sexo == 1) ], 
                                     base["Desercion de cursos"][(base.sexo == 0) ], equal_var=False)
print('t_value = %.3f, t_test_p = %.3f' % (t_value, t_test_p),"\n")

base.groupby(['sexo'])[['Desercion de cursos']].agg('mean').reset_index( level = ['sexo'])
```

```python
def show_values(axs, orient="v", space=.01):
    def _single(ax):
        if orient == "v":
            for p in ax.patches:
                _x = p.get_x() + p.get_width() / 2
                _y = p.get_y() + p.get_height() + (p.get_height()*0.01)
                value = '{:.1f}'.format(p.get_height())
                ax.text(_x, _y, value, ha="center") 
        elif orient == "h":
            for p in ax.patches:
                _x = p.get_x() + p.get_width() + float(space)
                _y = p.get_y() + p.get_height() - (p.get_height()*0.5)
                value = '{:.1f}'.format(p.get_width())
                ax.text(_x, _y, value, ha="left")

    if isinstance(axs, np.ndarray):
        for idx, ax in np.ndenumerate(axs):
            _single(ax)
    else:
        _single(axs)
```
