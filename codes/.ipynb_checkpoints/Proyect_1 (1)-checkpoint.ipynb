```python
import camelot as cm
import numpy as np
import pandas as pd
from pandas import Series, DataFrame
import xlrd
```

```python
file = 'Educacion_base.xlsx'

base_inicial = xlrd.open_workbook(file, on_demand=True)
sheet_names = base_inicial.sheet_names()

sheet_names
```

```python
base = pd.ExcelWriter('Educacion_base.xlsx')

for name in sheet_names:
 
    globals()[f'data_{name}'] = pd.read_excel(base, sheet_name= f'{name}')
```

## **Bucle final data**

```python
for name in sheet_names:

  if globals()[f'data_{name}'].iloc[1,1] == "NombredelIEST":
    globals()[f'data_{name}'] = globals()[f'data_{name}'].iloc[9:]

    globals()[f'data_{name}'] = globals()[f'data_{name}'].drop(globals()[f'data_{name}'].columns[0:2], axis = 1)
    

    globals()[f'data_{name}'][1][9] = "dni"
    globals()[f'data_{name}'][2][9] = "nombre"

    a = globals()[f'data_{name}'].iloc[1,-1]
    b = globals()[f'data_{name}'].iloc[1,-2]  

    globals()[f'data_{name}'] = globals()[f'data_{name}'].dropna(how='all', axis = 1)

    globals()[f'data_{name}'].iloc[:, -2] = globals()[f'data_{name}'].iloc[:, -2].fillna(b, inplace=False)
    globals()[f'data_{name}'].iloc[:, -1] = globals()[f'data_{name}'].iloc[:, -1].fillna(a, inplace=False)

    globals()[f'data_{name}'] = globals()[f'data_{name}'].drop(labels=10, axis = 0)
    globals()[f'data_{name}'] = globals()[f'data_{name}'].rename(columns=globals()[f'data_{name}'].iloc[0])

    globals()[f'data_{name}'] = globals()[f'data_{name}'].iloc[1:]
    globals()[f'data_{name}'] = globals()[f'data_{name}'].drop(['obs'] , axis = 1) 

    globals()[f'data_{name}'] = globals()[f'data_{name}'].dropna() 

  elif  globals()[f'data_{name}'].iloc[0,1] == "N°":
      globals()[f'data_{name}'] =  globals()[f'data_{name}'].iloc[3:]

      globals()[f'data_{name}'] = globals()[f'data_{name}'].drop(globals()[f'data_{name}'].columns[0:2], axis = 1)
      

      globals()[f'data_{name}'][1][3] = "dni"
      globals()[f'data_{name}'][2][3] = "nombre"
      
      a = globals()[f'data_{name}'].iloc[1,-1]
      b = globals()[f'data_{name}'].iloc[1,-2]

      globals()[f'data_{name}'] = globals()[f'data_{name}'].dropna(how='all', axis = 1)

      globals()[f'data_{name}'].iloc[:, -2] = globals()[f'data_{name}'].iloc[:, -2].fillna(b, inplace=False)
      globals()[f'data_{name}'].iloc[:, -1] = globals()[f'data_{name}'].iloc[:, -1].fillna(a, inplace=False)

      globals()[f'data_{name}'] = globals()[f'data_{name}'].drop(labels=4, axis = 0)
      globals()[f'data_{name}'] = globals()[f'data_{name}'].rename(columns=globals()[f'data_{name}'].iloc[0])
      globals()[f'data_{name}'] = globals()[f'data_{name}'].iloc[1:]
      globals()[f'data_{name}'] = globals()[f'data_{name}'].drop(['obs'] , axis = 1)  

      globals()[f'data_{name}'] = globals()[f'data_{name}'].dropna() 
```

```python
fc = pd.Series(base_completa.iloc[2290,2:229])
gc = pd.Series(base_completa.iloc[2290,229:268])

fc = fc.fillna(0)
gc = gc.fillna(0)

base_completa = base_completa.drop([10000])
base_completa
```

```python
ab = base_completa.iloc[:,2:268]
ab1 = ab.replace(np.nan, 0)

ab2 = ab.replace( np.arange(0.0,21.0), 1)
ab2 = ab2.replace(np.nan, 0)


```

```python
ab1['w_fc'] = ab1.iloc[:,:227].dot(fc)
ab1['w_gc'] = ab1.iloc[:,227:266].dot(gc)

ab2['t_cre_fc'] = ab2.iloc[:,:227].dot(fc)
ab2['t_cre_gc'] = ab2.iloc[:,227:266].dot(gc)
```

```python
ab1 = ab1[['w_fc', 'w_gc']]
ab2 = ab2[['t_cre_fc', 't_cre_gc']]
```

```python
base_completa = pd.concat([base_completa, ab1, ab2], axis=1)
```

```python
base_completa['ave_gc'] = base_completa['w_gc']/base_completa['t_cre_gc']
base_completa['ave_fc'] = base_completa['w_fc']/base_completa['t_cre_fc']
base_completa['ave_total'] = (base_completa['w_gc']+base_completa['w_fc'])/(base_completa['t_cre_fc']+base_completa['t_cre_gc'])
```
