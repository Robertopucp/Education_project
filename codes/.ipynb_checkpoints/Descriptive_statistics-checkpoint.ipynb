```python
def show_values(axs, orient="v", space=.01):
    def _single(ax):
        if orient == "v":
            for p in ax.patches:
                _x = p.get_x() + p.get_width() / 2
                _y = p.get_y() + p.get_height() + (p.get_height()*0.01)
                value = '{:.1f}'.format(p.get_height())
                ax.text(_x, _y, value, ha="center") 
        elif orient == "h":
            for p in ax.patches:
                _x = p.get_x() + p.get_width() + float(space)
                _y = p.get_y() + p.get_height() - (p.get_height()*0.5)
                value = '{:.1f}'.format(p.get_width())
                ax.text(_x, _y, value, ha="left")

    if isinstance(axs, np.ndarray):
        for idx, ax in np.ndenumerate(axs):
            _single(ax)
    else:
        _single(axs)
```

```python
plt.figure(figsize=(6,5))
ax = sns.barplot(x="carrera", y="percent_disaproved(%)", data=base4, color = 'lightgray',
                order=base4.sort_values('percent_disaproved(%)').carrera)

careers1 = ['C1','C2','C3', 'C4','C5', 'C6', 'C7']
careers2 = ['Administración de empresa (C1)','Mecánica de producción (C2)','Computación e informática (C3)', 'Contabilidad (C4)', 
           'Electrotecnia Industrial (C5)', 'Secretariado ejecutivo (C6)', 'Mecánica automotriz (C7)']

(ax.set_xticklabels(careers1))
plt.xlabel('')
plt.ylabel('Desertion rate (%)', size=11)
plt.tight_layout()
plt.legend(title='Technical careers', loc='upper right', labels=careers2, bbox_to_anchor=(1.6, 0.7), ncol=1)
show_values(ax)

des_rate19 = [2.9,10.2, 3.31, 0.96, 3.42, 0,0]

j = 0
for i in des_rate19:
    ax.text(j,i+0.1,i, color = 'red', size = 10, ha='center')
    j = j +1

#"ch:.25"
```
