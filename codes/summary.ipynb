```python
import warnings
warnings.filterwarnings('ignore')

sex = pd.read_csv("../base_datos/sexo_dni.csv")
panel = pd.read_csv('../outcomes/base_completa_panel.csv')
base_educacion = pd.merge(panel, sex, how="left", on="dni")


base = base_educacion[~base_educacion['id'].isna()]
base = base.iloc[: :,1:269] # seleccionando columnas 
base['n_c_desaprovados'] = ((base > 0) & (base < 12.5)).sum(axis=1) # Numero de cursos desaprobados
base['desc_cursos'] = (base.iloc[: :,3:266] == 0).sum(axis=1) # numero de cursos (desercion)


base_educacion = pd.merge(base_educacion, base, how = 'left', on = ['dni','period'])

iid = ["dni", "period", "sexo" , "section", "carrera", "shift", "ave_total", "ave_gc", "ave_fc", "rate_approv", 'n_c_desaprovados', 'desc_cursos']
rel_data = base_educacion[iid]

name_col = {
    'period':'Periodo', 'section':'Nivel', 'carrera':'Carrera','shift':'Turno',
    'ave_total':'Nota promedio general', 'ave_gc':'Nota promedio de cursos generales',
    'ave_fc':'Nota promedio de cursos de carrera','rate_approv':'Ratio de aprobacion por alumno',
    "n_c_desaprovados": "Numero de cursos desaprobados", "desc_cursos": "Decercion de cursos"
    }

```

```python
dic_fem = {1:"Female", 0: "Male"}
dic_per = {1: "2019 - 2", 2: "2020 - 1"}

rel_data_cat = rel_data.copy()

rel_data_cat["sexo"] = rel_data_cat["sexo"].map(dic_fem)
rel_data_cat["period"] = rel_data_cat["period"].map(dic_per)
```

```python
def compare_two(column, main_row, row, values,  data, fun = np.mean):
    result = pd.pivot_table(
    data, values=values, index=[main_row, row], columns=column
    , aggfunc=fun, fill_value=0)
    
    return result
    # result

s = compare_two("sexo", "carrera", "period", ["ave_gc", "ave_fc", "ave_total"], rel_data_cat, [np.mean]).round(2)
# rel_data
s
```

```python
compare_two("sexo", "carrera", "period", ["n_c_desaprovados"], rel_data_cat, [np.mean, len]).round(2)
```
