```python
#selec columns of GC and FC
fc = pd.Series(base_completa.iloc[2290,2:229])
gc = pd.Series(base_completa.iloc[2290,229:268])
# fill nanyo 0
fc = fc.fillna(0)
gc = gc.fillna(0)

# drop credits row
base_completa = base_completa.drop([10000])
base_completa
```

```python
# select columns grade and fill nan to 0
ab = base_completa.iloc[:,2:268]
ab1 = ab.replace(np.nan, 0)

# replace grades to 1 
ab2 = ab.replace( np.arange(0.0,21.0), 1)
ab2 = ab2.replace(np.nan, 0)


```

```python
# sum of field courses grades * credits 
ab1['w_fc'] = ab1.iloc[:,:227].dot(fc)
# sum of general courses grades * credits 
ab1['w_gc'] = ab1.iloc[:,227:266].dot(gc)

# sum of field courses credits
ab2['t_cre_fc'] = ab2.iloc[:,:227].dot(fc)
# sum of general courses credits
ab2['t_cre_gc'] = ab2.iloc[:,227:266].dot(gc)
```

```python
#keep columns of sum grades * credits and total credits 

ab1 = ab1[['w_fc', 'w_gc']]
ab2 = ab2[['t_cre_fc', 't_cre_gc']]
```

```python
# add grades columns on base_completa
base_completa = pd.concat([base_completa, ab1, ab2], axis=1)
```

```python
# average grades per GC and FC
base_completa['ave_gc'] = base_completa['w_gc']/base_completa['t_cre_gc']
base_completa['ave_fc'] = base_completa['w_fc']/base_completa['t_cre_fc']
base_completa['ave_total'] = (base_completa['w_gc']+base_completa['w_fc'])/(base_completa['t_cre_fc']+base_completa['t_cre_gc'])
```
